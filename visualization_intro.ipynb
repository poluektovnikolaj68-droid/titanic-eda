```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')

df['Age'].hist()
plt.title('Рапределение возраста пассажиров')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

Этот график показывает распределение возраста пассажиров. Как мы видим, большинство людей на Титанике были возрастом от 17 до примерно 32 лет.

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title('Выжившие VS Погибшие')
plt.xticks([0, 1], ['Погиб', 'Выжил'], rotation=0)
plt.ylabel('Количество')
plt.show()
```

Как мы видим, погибло примерно 525 человек, а выжило только около 350.

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title('Выживаемость по полу')
plt.ylabel('Доля выживших')
plt.xticks(rotation=0)
plt.show()
```

Женщин выжило в 3 раза больше чем мужчин.

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title('Выживаемость по классу')
plt.ylabel('Доля выживших')
plt.xticks(rotation=0)
plt.show()
```

Выживаемость в 1-м классе составила более 60%, в то время как во втором и третьем классе примерно 47% и 22% соответственно.
